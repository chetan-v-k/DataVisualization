```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("AB_NYC_2019.csv")
```

```python
df = data.copy()
df
```

```python
df.dropna(inplace=True)    #removes the rows with null values
df.drop_duplicates(inplace=True)  #removes the duplicate values
```

```python
#changing the data the of the columns

df['id'] = df['id'].astype("str")
df["host_id"] = df["host_id"].astype("str")
df["latitude"] = df['latitude'].astype("str")
df["longitude"] = df["longitude"].astype("str")
```

```python
df.describe()
```

```python
df.nunique()
```

```python
df.info()
```

Categorical Data Analysis

```python
#"neighbourhood_group"  has 5 unique categorical data
df["neighbourhood_group"].value_counts(normalize= True)   #normalize gives the percentage of the neighbourhood_group has in the NYC
```

```python
df["room_type"].value_counts(normalize=True)
```

```python
#here we are the column name "count" to "no_of_hotels"  and resetting the index
df_n = df["neighbourhood"].value_counts().reset_index().rename(columns= {"count":"no_of_hotels"})
```

```python
df_n
```

```python
df_n[df_n['neighbourhood']>'1000']
```

Numerical Data Analysis

```python
#bins = 5 means all the price values are put into 5 ranges(intervals)
df['price'].value_counts(bins=10)
```

```python
#Creating our own bins(ranges/intervals)
bins = (0,50,100,200,500,2000,5000,10000)
df["price"].value_counts(bins =bins)
```

```python
df["price"].mean()
```

```python
df["price"].std()
```

```python
df['price'].skew()
# Positive Skew (Right Skewed):
#Tail is longer on the right side.
#Most data is clustered on the left (lower values).
#Mean > Median > Mode

#Negative Skew (Left Skewed):
#Tail is longer on the left side.
#Most data is clustered on the right (higher values).
#Mean < Median < Mode
```

```python
df["price"].kurtosis()
```
